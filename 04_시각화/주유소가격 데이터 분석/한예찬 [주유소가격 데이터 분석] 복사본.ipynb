```python
import numpy as np
```

```python


import pandas as pd

fh = pd.read_csv('data/2019년 상반기 주유소 판매가격.csv')



sh = pd.read_csv('data/2019년 하반기 주유소 판매가격.csv') 
```

```python
# 상반기, 하반기 주유소 가격 데이터 조회
fh
```

```python
sh
```

```python
# 컬럼명을 기준으로 수직으로 합치기
# 상반기, 하반기 DataFrame 합치기
y = pd.concat([fh,sh],  keys=['상반기','하반기'])
y
```

```python
# DataFrame 기본 정보 조회
y.info()
```

```python
# 상표 컬럼의 고유값 조회
y['상표'].unique()
y['상표'].value_counts()
```

```python
# 지역 컬럼의 고유값 조회
y['지역'].unique()
y['지역'].value_counts()
```

> ### str accessor
> - `Series.str.String메소드()`
> - 문자열 타입 Series에 String 메소드들을 사용하여 일괄처리 할 수 있도록 한다.
> - str accessor 메소드 : https://pandas.pydata.org/pandas-docs/stable/reference/series.html#api-series-str

> ### datetime으로 변환
> - `pd.to_datetime(변환할 Series, format="날짜형식")`
> - x 또는 y 축이 날짜 일 경우 datetime 타입으로 처리하는 것이 좋다.
> - format: 년월일의 형식을 strftime 형식문자열로 지정한다. 
>      - [strftime 형식](https://docs.python.org/3/library/datetime.html#strftime-and-strptime-format-codes)  
> - datetime을 변환하면 날짜만의 기능을 다룰 수 있다.
>     - dt accessor를 이용해 datetime의 속성값들을(년, 월, 일 등) 추출할 수 있다.
>     - https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.Series.dt.html


```python
# "서울 서초구"[3:] => 슬라이싱
# re.findall('\w$','서울 서초구') => 정규표현식
```

```python
# 지역 컬럼의 값을 이용하여 "구" 컬럼 생성 (지역에서 구만 추출)
y['구'] = y['지역'].apply(lambda x: x.split(' ')[1])
```

```python
y['구'] = y['지역'].str.slice(3) #문자열 slicing
```

```python
y
```

```python
# 기간 컬럼을 datetime 타입으로 변경
from datetime import datetime
y['기간']= y['기간'].astype('str')
y['기간']= y['기간'].apply(lambda _ : datetime.strptime(_,'%Y%m%d'))
y
```

```python
y.info()
```

```python
# 기간 컬럼을 이용해 "월", "일", "요일" 컬럼 생성
# 요일 - 0:월, 1:화, 2:수, 3:목, 4:금, 5:토, 6:일 로 변경
y['월'] = y['기간'].dt.month
y['일'] = y['기간'].dt.day
y['요일']  = y['기간'].dt.day_name()
y
```

```python
#전처리 완료 DataFrame 저장
```

```python
# 휘발유 가격이 가장 비싼 5개 행 조회

# 휘발유가 가장비싼 1개 행
# df['휘발유'].max(), df['휘발유'].idxmax()
# df.iloc[df['휘발유'].idmax()]


y.sort_values('휘발유',ascending = False).head(5)
```

```python
# 상표별 휘발유 평균가격 조회
상표별_휘발유평균 = y.groupby('상표')['휘발유'].mean().sort_values() #평균가격의 오름차순
```
