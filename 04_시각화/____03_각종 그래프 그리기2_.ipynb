# 4. 파이차트 그리기
## 4.1 파이차트
- 각 범주(Category)가 데이터에서 차지하는 비율을 나타내는데 사용
- `pie(x, labels)` 이용
    - x: 값 (값들을 100을 기준으로 비율을 계산해 크기 설정)
    - labels : 값들의 label
    - autopct: 조각내에 표시될 비율의 문자열 형식. '%fmt문자' 
        - fmt문자: f(실수), d(정수), %% (%)
            

```python
import matplotlib.pyplot as plt

fruits = ['사과', '배', '귤', '딸기']
counts = [30, 10, 50, 20]
```

```python
import numpy as np
np.array(counts)/np.sum(counts)
```

```python
a = 12.56789
b = 3000
c = 'abc'
f'a:{a}, b:{b}, c:{c}'
'a:{}, b:{}, c:{}'.format(a,b,c)
# data type에 맞춰서 형식문자열을 지정. %d: 정수, %f: 실수, %s: 문자열
'a:%.3f, b:%d%%, c:%s' % (a, b, c)
```

```python
plt.figure(figsize=(10,10))
plt.pie(counts, labels=fruits,   #수량, labels=대상
        autopct='%.2f%%', 
        explode=[0.1,0,0.1,0], shadow=True)

plt.show()
```

# 5 히스토그램 그리기
## .1 히스토그램
- 표로 된 도수 분포표를 그래프로 나타낸 것.
    - 도수분포표: 특정 그룹안에 값의 빈도를 나타낸 표
    - 주로 연속형 자료를 특정 구간으로 나눠 그 빈도를 확인한다.
        - 빈도 확인이나 **분포**를 볼때 사용한다.
    - X축: 계급(변수구간)의 대표값, Y축: 빈도수
- `hist(data [, bins=계급개수)` 메소드를 사용
    - data: 리스트형의 객체를 전달한다.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 평균이 100이고 표준편차가 10인 정규분포를 따르는 임의수(난수-random값) 1000개를 생성
data = np.random.normal(100, 10, 1000)
print(data.shape)
data[:5], data[20:25], data[-5:]
```

```python
data_s = pd.Series(data)  #ndarray를 series 변환
v = data_s.value_counts().sort_values(ascending=False)
v
```

```python
# 연속형 값들을 수량을 센 뒤에 막대그래프로 시각화
plt.bar(v.index, v)
plt.show()
```

```python
plt.figure(figsize=(10,8))
plt.hist(data_s, bins=50)
plt.xlabel('bins')
plt.ylabel('수량')
plt.show()
```

# 6. 상자그래프(Boxplot) 그리기
## 6.1 상자그래프란
- 연속형 값들의 분포, 이상치(극단값)를 확인할 때 사용
    - 값들의 중앙값(중위수), 사분위수, 최대/최소값을 보여주어 값들의 분포를 확인 할때 사용한다.
- boxplot(x, whis=1.5)
    - x: 데이터셋
    - whis: 극단치 계산시 사용할 값(기본 1.5)
        - 극단치 : 1사분위와 3사분위에서 $IQR(3사분위수-1사분위수) * whis$ 범위 이상 떨어진 값들

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.Series(np.random.randint(1, 100, 1000))  # 1 ~ 100-1 사이의 임의의 정수 1000개를 생성
data.shape
```

```python
data.agg(["min", "max"])
```

```python
# 일부의 값을 극단적으로 크거나 작은 값으로 변경
# 극단치: 대부분의 값들이 분포된 범위에서 크게 떨어져 있는 값.
data[:5]  = [300, 290, -50, -40, 350]
data[:10]
```

```python
data.agg(['min', 'max'])
```

```python
plt.figure(figsize=(10,10))
plt.boxplot(data)
plt.ylabel('data값을 구성하는 값들')
plt.show()
```

- 연속형 값들의 분포 확인 - 히스토그램, boxplot
    - histogram: 연속된 값들을 그룹으로 나눈 뒤 수량을 시각화
    - boxplot:   값의 범위를 기준으로 시각확 (4분위수를 중심으로) => 수량정보는 없다.
